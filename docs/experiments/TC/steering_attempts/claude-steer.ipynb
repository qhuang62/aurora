The "steering_flow" perturbation creates an asymmetric wind pattern in the upper atmosphere (200-500 hPa levels) around the hurricane:

- North of storm: +5 m/s WESTWARD wind (u-component)
- Storm center:   0 m/s (no change)
- South of storm: +5 m/s EASTWARD wind (u-component)

This creates a north-south wind shear pattern that should theoretically push the hurricane northward by:

- The westward flow north of the storm creates a pressure gradient
- The eastward flow south creates an opposite gradient
- The hurricane should move toward the area of lower pressure (northward)

Additionally, a small northward velocity component (+1.5 m/s) is added across the region.

```python
import numpy as np
import torch
import xarray as xr
import matplotlib.pyplot as plt
from pathlib import Path
from aurora import Batch, Metadata
from datetime import datetime
from aurora import Aurora, Tracker, rollout

# ---------------- SETUP ----------------
download_path = Path("~/downloads").expanduser()
day = "2022-09-17"
static_vars_ds = xr.open_dataset(download_path / "static.nc", engine="netcdf4")
surf_vars_ds = xr.open_dataset(download_path / f"{day}-surface-level.nc", engine="netcdf4")
atmos_vars_ds = xr.open_dataset(download_path / f"{day}-atmospheric.nc", engine="netcdf4")

def _prepare(x: np.ndarray) -> torch.Tensor:
    return torch.from_numpy(x[[1, 2]][None][..., ::-1, :].copy())  # pick 06:00 and 12:00

# ---------------- BASELINE BATCH ----------------
u_baseline = atmos_vars_ds["u_component_of_wind"].values.copy()
v_baseline = atmos_vars_ds["v_component_of_wind"].values.copy()

batch_original = Batch(
    surf_vars={
        "2t": _prepare(surf_vars_ds["2m_temperature"].values),
        "10u": _prepare(surf_vars_ds["10m_u_component_of_wind"].values),
        "10v": _prepare(surf_vars_ds["10m_v_component_of_wind"].values),
        "msl": _prepare(surf_vars_ds["mean_sea_level_pressure"].values),
    },
    static_vars={
        "z": torch.from_numpy(static_vars_ds["z"].values[0]),
        "slt": torch.from_numpy(static_vars_ds["slt"].values[0]),
        "lsm": torch.from_numpy(static_vars_ds["lsm"].values[0]),
    },
    atmos_vars={
        "t": _prepare(atmos_vars_ds["temperature"].values),
        "u": _prepare(u_baseline),
        "v": _prepare(v_baseline),
        "q": _prepare(atmos_vars_ds["specific_humidity"].values),
        "z": _prepare(atmos_vars_ds["geopotential"].values),
    },
    metadata=Metadata(
        lat=torch.from_numpy(surf_vars_ds.latitude.values[::-1].copy()),
        lon=torch.from_numpy(surf_vars_ds.longitude.values),
        time=(surf_vars_ds.time.values.astype("datetime64[s]").tolist()[2],),
        atmos_levels=tuple(int(level) for level in atmos_vars_ds.level.values),
    ),
)

# ---------------- IMPROVED PERTURBATION STRATEGY ----------------
def create_perturbation_batch(strategy="steering_flow", magnitude=5.0):
    """
    Create perturbed batch with different strategies for hurricane steering
    
    Args:
        strategy: 'steering_flow', 'temperature_gradient', 'pressure_field', or 'combined'
        magnitude: perturbation strength
    """
    u_pert = atmos_vars_ds["u_component_of_wind"].values.copy()
    v_pert = atmos_vars_ds["v_component_of_wind"].values.copy()
    t_pert = atmos_vars_ds["temperature"].values.copy()
    msl_pert = surf_vars_ds["mean_sea_level_pressure"].values.copy()
    
    # Hurricane center and surrounding region
    lat_vals = atmos_vars_ds.latitude.values
    lon_vals = atmos_vars_ds.longitude.values
    lat_center = 27.5
    lon_center = 132.0
    
    lat_idx = np.argmin(np.abs(lat_vals - lat_center))
    lon_idx = np.argmin(np.abs(lon_vals - lon_center))
    
    # Define perturbation region (larger area for steering flow)
    lat_range = slice(lat_idx-8, lat_idx+8)  # ~200km radius
    lon_range = slice(lon_idx-8, lon_idx+8)
    
    if strategy == "steering_flow":
        # Modify upper-level steering flow (200-500 hPa)
        upper_levels = [0, 1, 2, 3]  # Assuming these correspond to upper levels
        
        # Create asymmetric flow pattern to steer hurricane
        for t_idx in [1, 2]:  # Both time steps
            for level_idx in upper_levels:
                # Add westward component north of storm, eastward south of storm
                lat_weights = np.linspace(1, -1, 16)  # From +1 (north) to -1 (south)
                lon_weights = np.ones(16)
                
                for i, lat_w in enumerate(lat_weights):
                    for j, lon_w in enumerate(lon_weights):
                        if lat_idx-8+i >= 0 and lat_idx-8+i < len(lat_vals) and \
                           lon_idx-8+j >= 0 and lon_idx-8+j < len(lon_vals):
                            # Steer northward by adding westward flow north, eastward flow south
                            u_pert[t_idx, level_idx, lat_idx-8+i, lon_idx-8+j] += magnitude * lat_w
                            # Add slight meridional component
                            v_pert[t_idx, level_idx, lat_idx-8+i, lon_idx-8+j] += magnitude * 0.3
    
    elif strategy == "temperature_gradient":
        # Modify temperature gradient to affect baroclinic environment
        mid_levels = [3, 4, 5, 6]  # Mid-troposphere
        
        for t_idx in [1, 2]:
            for level_idx in mid_levels:
                # Create north-south temperature gradient
                lat_gradient = np.linspace(-2, 2, 16)  # Cooler north, warmer south
                for i, temp_change in enumerate(lat_gradient):
                    for j in range(16):
                        if lat_idx-8+i >= 0 and lat_idx-8+i < t_pert.shape[2] and \
                           lon_idx-8+j >= 0 and lon_idx-8+j < t_pert.shape[3]:
                            t_pert[t_idx, level_idx, lat_idx-8+i, lon_idx-8+j] += temp_change
    
    elif strategy == "pressure_field":
        # Modify surface pressure field
        for t_idx in [1, 2]:
            # Create pressure dipole - higher pressure north, lower south
            lat_pressure = np.linspace(200, -200, 16)  # Pa
            for i, p_change in enumerate(lat_pressure):
                for j in range(16):
                    if lat_idx-8+i >= 0 and lat_idx-8+i < msl_pert.shape[1] and \
                       lon_idx-8+j >= 0 and lon_idx-8+j < msl_pert.shape[2]:
                        msl_pert[t_idx, lat_idx-8+i, lon_idx-8+j] += p_change
    
    elif strategy == "combined":
        # Apply multiple perturbations with reduced magnitude
        reduced_mag = magnitude * 0.5
        
        # Steering flow
        upper_levels = [0, 1, 2]
        for t_idx in [1, 2]:
            for level_idx in upper_levels:
                lat_weights = np.linspace(1, -1, 16)
                for i, lat_w in enumerate(lat_weights):
                    for j in range(16):
                        if lat_idx-8+i >= 0 and lat_idx-8+i < len(lat_vals) and \
                           lon_idx-8+j >= 0 and lon_idx-8+j < len(lon_vals):
                            u_pert[t_idx, level_idx, lat_idx-8+i, lon_idx-8+j] += reduced_mag * lat_w
                            v_pert[t_idx, level_idx, lat_idx-8+i, lon_idx-8+j] += reduced_mag * 0.2
        
        # Temperature gradient
        mid_levels = [4, 5]
        for t_idx in [1, 2]:
            for level_idx in mid_levels:
                lat_gradient = np.linspace(-1, 1, 16)
                for i, temp_change in enumerate(lat_gradient):
                    for j in range(16):
                        if lat_idx-8+i >= 0 and lat_idx-8+i < t_pert.shape[2] and \
                           lon_idx-8+j >= 0 and lon_idx-8+j < t_pert.shape[3]:
                            t_pert[t_idx, level_idx, lat_idx-8+i, lon_idx-8+j] += temp_change
    
    # Create perturbed batch
    batch_perturbed = Batch(
        surf_vars={
            "2t": _prepare(surf_vars_ds["2m_temperature"].values),
            "10u": _prepare(surf_vars_ds["10m_u_component_of_wind"].values),
            "10v": _prepare(surf_vars_ds["10m_v_component_of_wind"].values),
            "msl": _prepare(msl_pert),
        },
        static_vars={
            "z": torch.from_numpy(static_vars_ds["z"].values[0]),
            "slt": torch.from_numpy(static_vars_ds["slt"].values[0]),
            "lsm": torch.from_numpy(static_vars_ds["lsm"].values[0]),
        },
        atmos_vars={
            "t": _prepare(t_pert),
            "u": _prepare(u_pert),
            "v": _prepare(v_pert),
            "q": _prepare(atmos_vars_ds["specific_humidity"].values),
            "z": _prepare(atmos_vars_ds["geopotential"].values),
        },
        metadata=Metadata(
            lat=torch.from_numpy(surf_vars_ds.latitude.values[::-1].copy()),
            lon=torch.from_numpy(surf_vars_ds.longitude.values),
            time=(surf_vars_ds.time.values.astype("datetime64[s]").tolist()[2],),
            atmos_levels=tuple(int(level) for level in atmos_vars_ds.level.values),
        ),
    )
    
    return batch_perturbed

# ---------- IMPROVED TRACKING FUNCTION ----------
def run_and_compare_tracks(model, batch_original, batch_perturbed, steps=16):
    """Compare tracks with improved visualization and analysis"""
    tracker_orig = Tracker(init_lat=27.5, init_lon=132, init_time=datetime(2022, 9, 17, 12))
    tracker_pert = Tracker(init_lat=27.5, init_lon=132, init_time=datetime(2022, 9, 17, 12))
    
    preds_orig, preds_pert = [], []
    
    with torch.inference_mode():
        for pred_o, pred_p in zip(
            rollout(model, batch_original, steps=steps),
            rollout(model, batch_perturbed, steps=steps)
        ):
            preds_orig.append(pred_o.to("cpu"))
            preds_pert.append(pred_p.to("cpu"))
            tracker_orig.step(pred_o)
            tracker_pert.step(pred_p)
    
    # Extract results
    orig = tracker_orig.results()
    pert = tracker_pert.results()
    
    # Enhanced plotting
    plt.figure(figsize=(12, 8))
    
    plt.subplot(1, 2, 1)
    plt.plot(orig.lon, orig.lat, marker='o', label='Original', color='blue', linewidth=2, markersize=6)
    plt.plot(pert.lon, pert.lat, marker='x', label='Perturbed', color='red', linewidth=2, markersize=8)
    plt.title("Hurricane Track Comparison")
    plt.xlabel("Longitude")
    plt.ylabel("Latitude")
    plt.grid(True, alpha=0.3)
    plt.legend()
    
    # Add coastline context (rough Japan outline)
    japan_coast_lat = [35, 35, 31, 31, 35]
    japan_coast_lon = [135, 140, 140, 135, 135]
    plt.plot(japan_coast_lon, japan_coast_lat, 'k-', alpha=0.5, label='Japan coast (rough)')
    
    # Calculate and plot deviations
    plt.subplot(1, 2, 2)
    dist_km = []
    times = []
    
    for i in range(len(orig.lat)):
        # Great circle distance approximation
        dlat = np.radians(pert.lat[i] - orig.lat[i])
        dlon = np.radians(pert.lon[i] - orig.lon[i])
        lat_avg = np.radians((pert.lat[i] + orig.lat[i]) / 2)
        
        distance = 6371 * np.sqrt(dlat**2 + (dlon * np.cos(lat_avg))**2)
        dist_km.append(distance)
        times.append(i * 6)  # 6-hour intervals
    
    plt.plot(times, dist_km, 'g-o', linewidth=2, markersize=6)
    plt.title("Track Deviation Over Time")
    plt.xlabel("Forecast Hours")
    plt.ylabel("Distance Deviation (km)")
    plt.grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.show()
    
    # Print detailed results
    print(f"\nTrack Analysis:")
    print(f"Initial position: {orig.lat[0]:.2f}N, {orig.lon[0]:.2f}E")
    print(f"Final deviation: {dist_km[-1]:.1f} km")
    print(f"Maximum deviation: {max(dist_km):.1f} km at hour {times[np.argmax(dist_km)]}")
    
    print(f"\nStep-by-step deviations:")
    for i, (d, t) in enumerate(zip(dist_km, times)):
        direction = "N" if pert.lat[i] > orig.lat[i] else "S"
        ew_direction = "E" if pert.lon[i] > orig.lon[i] else "W"
        print(f"Hour {t:2d}: {d:6.1f} km deviation ({direction}{ew_direction})")
    
    return preds_orig, preds_pert, orig, pert, dist_km

# ---------- LOAD MODEL ----------
model = Aurora()
model.load_checkpoint("microsoft/aurora", "aurora-0.25-finetuned.ckpt")
model.eval()
model = model.to("cuda")

# ---------- TEST DIFFERENT STRATEGIES ----------
strategies = ["steering_flow", "temperature_gradient", "pressure_field", "combined"]
magnitudes = [3.0, 5.0, 8.0]

results = {}

for strategy in strategies:
    print(f"\n{'='*50}")
    print(f"Testing strategy: {strategy.upper()}")
    print(f"{'='*50}")
    
    for magnitude in magnitudes:
        print(f"\nMagnitude: {magnitude}")
        print("-" * 30)
        
        # Create perturbed batch
        batch_perturbed = create_perturbation_batch(strategy=strategy, magnitude=magnitude)
        
        # Run comparison
        try:
            preds_orig, preds_pert, track_orig, track_pert, deviations = run_and_compare_tracks(
                model, batch_original, batch_perturbed, steps=12
            )
            
            results[f"{strategy}_{magnitude}"] = {
                "final_deviation": deviations[-1],
                "max_deviation": max(deviations),
                "track_orig": track_orig,
                "track_pert": track_pert
            }
            
        except Exception as e:
            print(f"Error with {strategy} at magnitude {magnitude}: {e}")

# ---------- SUMMARY ----------
print(f"\n{'='*60}")
print("SUMMARY OF ALL EXPERIMENTS")
print(f"{'='*60}")

for key, result in results.items():
    strategy, magnitude = key.rsplit('_', 1)
    print(f"{strategy:20s} (mag={magnitude}): Final={result['final_deviation']:6.1f}km, Max={result['max_deviation']:6.1f}km")

model = model.to("cpu")  # Free GPU memory
```

```python
##not ran -- take too long
import numpy as np
import torch
import xarray as xr
import matplotlib.pyplot as plt
from pathlib import Path
from aurora import Batch, Metadata
from datetime import datetime
from aurora import Aurora, Tracker, rollout

# ---------------- SETUP ----------------
download_path = Path("~/downloads").expanduser()
day = "2022-09-17"
static_vars_ds = xr.open_dataset(download_path / "static.nc", engine="netcdf4")
surf_vars_ds = xr.open_dataset(download_path / f"{day}-surface-level.nc", engine="netcdf4")
atmos_vars_ds = xr.open_dataset(download_path / f"{day}-atmospheric.nc", engine="netcdf4")

def _prepare(x: np.ndarray) -> torch.Tensor:
    return torch.from_numpy(x[[1, 2]][None][..., ::-1, :].copy())  # pick 06:00 and 12:00

# ---------------- DIAGNOSTIC FUNCTION ----------------
def diagnose_data_structure():
    """Understand the data structure to fix indexing issues"""
    print("Data Structure Diagnosis:")
    print(f"Surface vars shape: {surf_vars_ds['2m_temperature'].shape}")
    print(f"Atmospheric vars shape: {atmos_vars_ds['temperature'].shape}")
    print(f"Time coordinates: {surf_vars_ds.time.values}")
    print(f"Pressure levels: {atmos_vars_ds.level.values}")
    print(f"Lat range: {atmos_vars_ds.latitude.values.min():.1f} to {atmos_vars_ds.latitude.values.max():.1f}")
    print(f"Lon range: {atmos_vars_ds.longitude.values.min():.1f} to {atmos_vars_ds.longitude.values.max():.1f}")
    
    # Find hurricane location in grid
    lat_vals = atmos_vars_ds.latitude.values
    lon_vals = atmos_vars_ds.longitude.values
    lat_idx = np.argmin(np.abs(lat_vals - 27.5))
    lon_idx = np.argmin(np.abs(lon_vals - 132))
    
    print(f"\nHurricane location:")
    print(f"Target: 27.5N, 132E")
    print(f"Nearest grid: {lat_vals[lat_idx]:.1f}N, {lon_vals[lon_idx]:.1f}E")
    print(f"Grid indices: lat={lat_idx}, lon={lon_idx}")
    
    return lat_idx, lon_idx

# Run diagnosis
lat_center_idx, lon_center_idx = diagnose_data_structure()

# ---------------- BASELINE BATCH ----------------
batch_original = Batch(
    surf_vars={
        "2t": _prepare(surf_vars_ds["2m_temperature"].values),
        "10u": _prepare(surf_vars_ds["10m_u_component_of_wind"].values),  
        "10v": _prepare(surf_vars_ds["10m_v_component_of_wind"].values),
        "msl": _prepare(surf_vars_ds["mean_sea_level_pressure"].values),
    },
    static_vars={
        "z": torch.from_numpy(static_vars_ds["z"].values[0]),
        "slt": torch.from_numpy(static_vars_ds["slt"].values[0]), 
        "lsm": torch.from_numpy(static_vars_ds["lsm"].values[0]),
    },
    atmos_vars={
        "t": _prepare(atmos_vars_ds["temperature"].values),
        "u": _prepare(atmos_vars_ds["u_component_of_wind"].values),
        "v": _prepare(atmos_vars_ds["v_component_of_wind"].values),
        "q": _prepare(atmos_vars_ds["specific_humidity"].values),
        "z": _prepare(atmos_vars_ds["geopotential"].values),
    },
    metadata=Metadata(
        lat=torch.from_numpy(surf_vars_ds.latitude.values[::-1].copy()),
        lon=torch.from_numpy(surf_vars_ds.longitude.values),
        time=(surf_vars_ds.time.values.astype("datetime64[s]").tolist()[2],),
        atmos_levels=tuple(int(level) for level in atmos_vars_ds.level.values),
    ),
)

# ---------------- IMPROVED PERTURBATION WITH DEBUGGING ----------------
def create_strong_perturbation_batch(strategy="vortex_relocation", magnitude=15.0):
    """
    Create strong, physically-based perturbations that should definitely affect tracks
    """
    # Copy all data
    u_pert = atmos_vars_ds["u_component_of_wind"].values.copy()
    v_pert = atmos_vars_ds["v_component_of_wind"].values.copy()
    t_pert = atmos_vars_ds["temperature"].values.copy()
    msl_pert = surf_vars_ds["mean_sea_level_pressure"].values.copy()
    surf_u_pert = surf_vars_ds["10m_u_component_of_wind"].values.copy()
    surf_v_pert = surf_vars_ds["10m_v_component_of_wind"].values.copy()
    
    # Get dimensions
    n_times, n_levels, n_lats, n_lons = u_pert.shape
    n_surf_times, n_surf_lats, n_surf_lons = msl_pert.shape
    
    print(f"\nApplying {strategy} perturbation with magnitude {magnitude}")
    print(f"Atmospheric shape: {u_pert.shape}")
    print(f"Surface shape: {msl_pert.shape}")
    
    # Define large perturbation region
    lat_radius = 15  # ~300km radius  
    lon_radius = 15
    
    lat_start = max(0, lat_center_idx - lat_radius)
    lat_end = min(n_lats, lat_center_idx + lat_radius)
    lon_start = max(0, lon_center_idx - lon_radius) 
    lon_end = min(n_lons, lon_center_idx + lon_radius)
    
    print(f"Perturbation region: lat[{lat_start}:{lat_end}], lon[{lon_start}:{lon_end}]")
    
    if strategy == "vortex_relocation":
        # Simulate moving the hurricane center by creating strong asymmetric flow
        print("Creating vortex relocation pattern...")
        
        # Apply to multiple time steps and levels
        for t_idx in [1, 2] if n_times > 2 else range(n_times):
            # Upper levels (steering flow) - use first few levels
            for level_idx in range(min(4, n_levels)):
                for i in range(lat_start, lat_end):
                    for j in range(lon_start, lon_end):
                        # Distance from hurricane center
                        di = i - lat_center_idx
                        dj = j - lon_center_idx
                        dist = np.sqrt(di**2 + dj**2)
                        
                        if dist < lat_radius:
                            # Create northward steering by adding westward flow north of center
                            weight = np.exp(-dist/8) * magnitude  # Gaussian weighting
                            
                            # North of center: add westward (negative u) component
                            if di < 0:  # North (remember lat is flipped)
                                u_pert[t_idx, level_idx, i, j] -= weight
                                v_pert[t_idx, level_idx, i, j] += weight * 0.3
                            # South of center: add eastward (positive u) component  
                            else:
                                u_pert[t_idx, level_idx, i, j] += weight
                                v_pert[t_idx, level_idx, i, j] += weight * 0.3
    
    elif strategy == "pressure_steering":
        print("Creating pressure steering pattern...")
        
        # Strong surface pressure anomalies
        for t_idx in [1, 2] if n_surf_times > 2 else range(n_surf_times):
            for i in range(lat_start, lat_end):
                for j in range(lon_start, lon_end):
                    di = i - lat_center_idx
                    dj = j - lon_center_idx
                    dist = np.sqrt(di**2 + dj**2)
                    
                    if dist < lat_radius:
                        weight = np.exp(-dist/10)
                        # High pressure north, low pressure south
                        if di < 0:  # North
                            msl_pert[t_idx, i, j] += magnitude * 100 * weight  # 100 Pa per unit
                        else:  # South
                            msl_pert[t_idx, i, j] -= magnitude * 100 * weight
    
    elif strategy == "temperature_steering":
        print("Creating temperature steering pattern...")
        
        # Strong temperature gradients in mid-troposphere
        mid_levels = range(n_levels//3, 2*n_levels//3)  # Middle third of levels
        
        for t_idx in [1, 2] if n_times > 2 else range(n_times):
            for level_idx in mid_levels:
                for i in range(lat_start, lat_end):
                    for j in range(lon_start, lon_end):
                        di = i - lat_center_idx
                        dj = j - lon_center_idx
                        dist = np.sqrt(di**2 + dj**2)
                        
                        if dist < lat_radius:
                            weight = np.exp(-dist/8)
                            # Warm north, cool south (or vice versa)
                            if di < 0:  # North
                                t_pert[t_idx, level_idx, i, j] += magnitude * weight
                            else:  # South
                                t_pert[t_idx, level_idx, i, j] -= magnitude * weight
    
    elif strategy == "extreme_combined":
        print("Creating extreme combined perturbation...")
        
        # All perturbations at once with high magnitude
        for t_idx in [1, 2] if n_times > 2 else range(n_times):
            # Wind perturbations at all levels
            for level_idx in range(n_levels):
                for i in range(lat_start, lat_end):
                    for j in range(lon_start, lon_end):
                        di = i - lat_center_idx
                        dj = j - lon_center_idx
                        dist = np.sqrt(di**2 + dj**2)
                        
                        if dist < lat_radius:
                            weight = np.exp(-dist/6) * magnitude
                            
                            # Strong asymmetric flow
                            if di < 0:  # North
                                u_pert[t_idx, level_idx, i, j] -= weight
                                v_pert[t_idx, level_idx, i, j] += weight * 0.5
                                if level_idx in range(n_levels//3, 2*n_levels//3):
                                    t_pert[t_idx, level_idx, i, j] += weight * 0.5
                            else:  # South
                                u_pert[t_idx, level_idx, i, j] += weight
                                v_pert[t_idx, level_idx, i, j] += weight * 0.5
                                if level_idx in range(n_levels//3, 2*n_levels//3):
                                    t_pert[t_idx, level_idx, i, j] -= weight * 0.5
            
            # Surface perturbations
            if t_idx < n_surf_times:
                for i in range(lat_start, lat_end):
                    for j in range(lon_start, lon_end):
                        di = i - lat_center_idx
                        dj = j - lon_center_idx
                        dist = np.sqrt(di**2 + dj**2)
                        
                        if dist < lat_radius:
                            weight = np.exp(-dist/8)
                            if di < 0:  # North
                                msl_pert[t_idx, i, j] += magnitude * 200 * weight
                                surf_u_pert[t_idx, i, j] -= magnitude * weight
                                surf_v_pert[t_idx, i, j] += magnitude * 0.3 * weight
                            else:  # South
                                msl_pert[t_idx, i, j] -= magnitude * 200 * weight
                                surf_u_pert[t_idx, i, j] += magnitude * weight
                                surf_v_pert[t_idx, i, j] += magnitude * 0.3 * weight
    
    # Print perturbation statistics
    original_u = atmos_vars_ds["u_component_of_wind"].values
    u_diff = np.abs(u_pert - original_u).max()
    v_diff = np.abs(v_pert - atmos_vars_ds["v_component_of_wind"].values).max()
    t_diff = np.abs(t_pert - atmos_vars_ds["temperature"].values).max()
    p_diff = np.abs(msl_pert - surf_vars_ds["mean_sea_level_pressure"].values).max()
    
    print(f"Maximum perturbations applied:")
    print(f"  U-wind: {u_diff:.2f} m/s")
    print(f"  V-wind: {v_diff:.2f} m/s") 
    print(f"  Temperature: {t_diff:.2f} K")
    print(f"  Pressure: {p_diff:.2f} Pa")
    
    # Create batch with all perturbations
    batch_perturbed = Batch(
        surf_vars={
            "2t": _prepare(surf_vars_ds["2m_temperature"].values),
            "10u": _prepare(surf_u_pert),
            "10v": _prepare(surf_v_pert),
            "msl": _prepare(msl_pert),
        },
        static_vars={
            "z": torch.from_numpy(static_vars_ds["z"].values[0]),
            "slt": torch.from_numpy(static_vars_ds["slt"].values[0]),
            "lsm": torch.from_numpy(static_vars_ds["lsm"].values[0]),
        },
        atmos_vars={
            "t": _prepare(t_pert),
            "u": _prepare(u_pert), 
            "v": _prepare(v_pert),
            "q": _prepare(atmos_vars_ds["specific_humidity"].values),
            "z": _prepare(atmos_vars_ds["geopotential"].values),
        },
        metadata=Metadata(
            lat=torch.from_numpy(surf_vars_ds.latitude.values[::-1].copy()),
            lon=torch.from_numpy(surf_vars_ds.longitude.values),
            time=(surf_vars_ds.time.values.astype("datetime64[s]").tolist()[2],),
            atmos_levels=tuple(int(level) for level in atmos_vars_ds.level.values),
        ),
    )
    
    return batch_perturbed

# ---------- ENHANCED TRACKING FUNCTION ----------
def run_detailed_track_comparison(model, batch_original, batch_perturbed, steps=20):
    """Enhanced tracking with better precision and analysis"""
    
    tracker_orig = Tracker(init_lat=27.5, init_lon=132.0, init_time=datetime(2022, 9, 17, 12))
    tracker_pert = Tracker(init_lat=27.5, init_lon=132.0, init_time=datetime(2022, 9, 17, 12))
    
    preds_orig, preds_pert = [], []
    
    print(f"Running {steps}-step forecast...")
    
    with torch.inference_mode():
        orig_rollout = rollout(model, batch_original, steps=steps)
        pert_rollout = rollout(model, batch_perturbed, steps=steps)
        
        for step, (pred_o, pred_p) in enumerate(zip(orig_rollout, pert_rollout)):
            preds_orig.append(pred_o.to("cpu"))
            preds_pert.append(pred_p.to("cpu"))
            
            tracker_orig.step(pred_o)
            tracker_pert.step(pred_p)
            
            # Print intermediate positions
            if step % 4 == 0:  # Every 24 hours
                orig_pos = tracker_orig.results()
                pert_pos = tracker_pert.results()
                if len(orig_pos.lat) > 0 and len(pert_pos.lat) > 0:
                    print(f"  Step {step+1}: Original({orig_pos.lat[-1]:.2f}, {orig_pos.lon[-1]:.2f}) "
                          f"Perturbed({pert_pos.lat[-1]:.2f}, {pert_pos.lon[-1]:.2f})")
    
    # Extract results
    orig = tracker_orig.results()
    pert = tracker_pert.results()
    
    # Calculate precise distances using haversine formula
    def haversine_distance(lat1, lon1, lat2, lon2):
        R = 6371  # Earth radius in km
        dlat = np.radians(lat2 - lat1)
        dlon = np.radians(lon2 - lon1)
        a = np.sin(dlat/2)**2 + np.cos(np.radians(lat1)) * np.cos(np.radians(lat2)) * np.sin(dlon/2)**2
        c = 2 * np.arcsin(np.sqrt(a))
        return R * c
    
    distances = []
    for i in range(min(len(orig.lat), len(pert.lat))):
        dist = haversine_distance(orig.lat[i], orig.lon[i], pert.lat[i], pert.lon[i])
        distances.append(dist)
    
    # Enhanced plotting
    fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(15, 12))
    
    # Track comparison
    ax1.plot(orig.lon, orig.lat, 'b-o', label='Original', linewidth=3, markersize=8)
    ax1.plot(pert.lon, pert.lat, 'r-x', label='Perturbed', linewidth=3, markersize=10)
    ax1.set_title("Hurricane Track Comparison", fontsize=14)
    ax1.set_xlabel("Longitude")
    ax1.set_ylabel("Latitude")
    ax1.grid(True, alpha=0.3)
    ax1.legend()
    
    # Distance evolution
    times = np.arange(len(distances)) * 6  # 6-hour intervals
    ax2.plot(times, distances, 'g-o', linewidth=3, markersize=6)
    ax2.set_title("Track Deviation Over Time", fontsize=14)
    ax2.set_xlabel("Forecast Hours")
    ax2.set_ylabel("Distance Deviation (km)")
    ax2.grid(True, alpha=0.3)
    
    # Speed comparison
    if len(orig.lat) > 1:
        orig_speeds = []
        pert_speeds = []
        for i in range(1, min(len(orig.lat), len(pert.lat))):
            orig_speed = haversine_distance(orig.lat[i-1], orig.lon[i-1], orig.lat[i], orig.lon[i]) / 6
            pert_speed = haversine_distance(pert.lat[i-1], pert.lon[i-1], pert.lat[i], pert.lon[i]) / 6
            orig_speeds.append(orig_speed)
            pert_speeds.append(pert_speed)
        
        ax3.plot(times[1:len(orig_speeds)+1], orig_speeds, 'b-', label='Original', linewidth=2)
        ax3.plot(times[1:len(pert_speeds)+1], pert_speeds, 'r-', label='Perturbed', linewidth=2)
        ax3.set_title("Hurricane Speed", fontsize=14)
        ax3.set_xlabel("Forecast Hours")
        ax3.set_ylabel("Speed (km/h)")
        ax3.grid(True, alpha=0.3)
        ax3.legend()
    
    # Direction comparison
    if len(orig.lat) > 1:
        orig_dirs = []
        pert_dirs = []
        for i in range(1, min(len(orig.lat), len(pert.lat))):
            orig_dir = np.degrees(np.arctan2(orig.lat[i] - orig.lat[i-1], orig.lon[i] - orig.lon[i-1]))
            pert_dir = np.degrees(np.arctan2(pert.lat[i] - pert.lat[i-1], pert.lon[i] - pert.lon[i-1]))
            orig_dirs.append(orig_dir)
            pert_dirs.append(pert_dir)
        
        ax4.plot(times[1:len(orig_dirs)+1], orig_dirs, 'b-', label='Original', linewidth=2)
        ax4.plot(times[1:len(pert_dirs)+1], pert_dirs, 'r-', label='Perturbed', linewidth=2)
        ax4.set_title("Hurricane Direction", fontsize=14)
        ax4.set_xlabel("Forecast Hours")
        ax4.set_ylabel("Direction (degrees)")
        ax4.grid(True, alpha=0.3)
        ax4.legend()
    
    plt.tight_layout()
    plt.show()
    
    # Detailed analysis
    if len(distances) > 0:
        print(f"\n{'='*60}")
        print("DETAILED TRACK ANALYSIS")
        print(f"{'='*60}")
        print(f"Number of forecast steps: {len(distances)}")
        print(f"Final deviation: {distances[-1]:.1f} km")
        print(f"Maximum deviation: {max(distances):.1f} km at hour {times[np.argmax(distances)]}")
        print(f"Average deviation: {np.mean(distances):.1f} km")
        print(f"Deviation growth rate: {(distances[-1] - distances[0]) / len(distances):.2f} km/step")
        
        # Print hourly deviations
        print(f"\nHourly track deviations:")
        for i, (d, t) in enumerate(zip(distances, times)):
            if i < len(orig.lat) and i < len(pert.lat):
                dlat = pert.lat[i] - orig.lat[i]
                dlon = pert.lon[i] - orig.lon[i]
                direction = f"{'N' if dlat > 0 else 'S'}{'E' if dlon > 0 else 'W'}"
                print(f"Hour {t:3d}: {d:7.1f} km ({direction}) - "
                      f"O:({orig.lat[i]:6.2f},{orig.lon[i]:7.2f}) "
                      f"P:({pert.lat[i]:6.2f},{pert.lon[i]:7.2f})")
    
    return preds_orig, preds_pert, orig, pert, distances

# ---------- LOAD MODEL ----------
print("Loading Aurora model...")
model = Aurora()
model.load_checkpoint("microsoft/aurora", "aurora-0.25-finetuned.ckpt")
model.eval()
model = model.to("cuda")

# ---------- TEST EXTREME PERTURBATIONS ----------
strategies = ["vortex_relocation", "pressure_steering", "temperature_steering", "extreme_combined"]
magnitudes = [10.0, 20.0, 50.0]  # Much larger magnitudes

results = {}

for strategy in strategies:
    print(f"\n{'='*70}")
    print(f"TESTING STRATEGY: {strategy.upper()}")
    print(f"{'='*70}")
    
    for magnitude in magnitudes:
        print(f"\n{'-'*50}")
        print(f"MAGNITUDE: {magnitude}")
        print(f"{'-'*50}")
        
        try:
            # Create strongly perturbed batch
            batch_perturbed = create_strong_perturbation_batch(strategy=strategy, magnitude=magnitude)
            
            # Run detailed comparison
            preds_orig, preds_pert, track_orig, track_pert, deviations = run_detailed_track_comparison(
                model, batch_original, batch_perturbed, steps=16
            )
            
            if len(deviations) > 0:
                results[f"{strategy}_{magnitude}"] = {
                    "final_deviation": deviations[-1],
                    "max_deviation": max(deviations),
                    "avg_deviation": np.mean(deviations),
                    "track_orig": track_orig,
                    "track_pert": track_pert
                }
            else:
                print("WARNING: No deviations calculated!")
                
        except Exception as e:
            print(f"ERROR with {strategy} at magnitude {magnitude}: {e}")
            import traceback
            traceback.print_exc()

# ---------- FINAL SUMMARY ----------
print(f"\n{'='*80}")
print("FINAL SUMMARY OF ALL EXTREME PERTURBATION EXPERIMENTS")
print(f"{'='*80}")

if results:
    for key, result in results.items():
        strategy, magnitude = key.rsplit('_', 1)
        print(f"{strategy:20s} (mag={magnitude:4s}): "
              f"Final={result['final_deviation']:7.1f}km, "
              f"Max={result['max_deviation']:7.1f}km, "
              f"Avg={result['avg_deviation']:7.1f}km")
    
    # Find best perturbation
    best_key = max(results.keys(), key=lambda k: results[k]['final_deviation'])
    best_result = results[best_key]
    strategy, magnitude = best_key.rsplit('_', 1)
    
    print(f"\nBEST PERTURBATION:")
    print(f"Strategy: {strategy} with magnitude {magnitude}")
    print(f"Final deviation: {best_result['final_deviation']:.1f} km")
    print(f"Maximum deviation: {best_result['max_deviation']:.1f} km")
    
else:
    print("No successful results obtained. Check data structure and model compatibility.")

model = model.to("cpu")  # Free GPU memory
print("\nAnalysis complete!")
```
